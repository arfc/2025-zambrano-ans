```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("EPS_data_nuclear.csv", index_col='index')
df
```

```python
#plotting the associated eps values against number of runs
ax = df.plot()
ax.minorticks_on()
ax.grid(which='both')
```

```python
#rolling mean on data to see changes between a specific generation and the preceding generation
ax = df.rolling(20).mean().plot()
ax.axhline(y=0.05, color='r', linestyle='--')
```
